```python
class Stack:
    
    def __init__(self):
        self.data = []
    
    def __len__(self):
        return len(self.data)
    
    def is_empty(self):
        return len(self.data) == 0
    
    def push(self, value):
        self.data.append(value)
        
    def top(self):
        return self.data[-1]
    
    def pop(self):
        if self.is_empty():
            print("Stack is Underflow")
        else:
            return self.data.pop()
```

# Queue

FIFO

```python
  30   40   50   60    70
 rare                 front
```

```python
[None] * 3
```

```python
class Queue:
    
    DEFAULT_CAPACITY = 10
    
    def __init__(self):
        self.data = [None] * Queue.DEFAULT_CAPACITY
        self.size = 0
        self.front = 0
        self.rare = 0
    
    def __len__(self):
        return self.size
    
    def is_empty(self):
        return self.size == 0
    
    def top(self):
        if self.is_empty():
            raise "Queue is Empty"
        return self.data[self.front]
    
    # dequeue - delete an element
    def dequeue(self):
        if self.is_empty():
            raise "Queue is empty"
        answer = self.data[self.rare+1]
        self.data[self.rare] = None
        self.rare += 1
        self.size += 1
        return answer
    
    # enqueue - add an element
    def enqueue(self, element):
        if self.size == Queue.DEFAULT_CAPACITY:
            raise "Queue is overflow"
        
#         a = self.front + 1
        self.data[self.front] = element
        self.front += 1
        self.size += 1
    
    # resize the queue
```

```python
q = Queue()
```

```python
q.enqueue(10)
```

```python
q.enqueue(20)
```

```python
print(q.data)
```
